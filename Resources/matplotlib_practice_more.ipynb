```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')  
# 'o' switches from a line chart to scatter in this format
```

```python
# OR

plt.scatter(y_axis, x_axis, c = 'red', label="chicago")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.legend()
```

```python
# create a bubble chart using the MATLAB method
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
# using list comprehension to replcace for and while loops of plots

# format == new_list = [expression for item in list if conditional]
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# scatter plot using the object oriented interface
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=(y_axis_larger), c='skyblue')




# CONTINUE WITH THIS CHALLENGE
```

```python
# PIE CHARTS::
```

```python
# y axis has the values (larger = larger slice)
# x axis has values (ea. slice)

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))  # fig(width,height)
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')  # % is necessary both before and after 

plt.show()
```
