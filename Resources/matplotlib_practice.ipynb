```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# The pyplot module will allow us to create all the necessary charts; create a plotting area in a figure; 
# plot lines in a plotting area; and annotate the chart with labels, colors, and other decorative features.
```

```python
# Set the x-axis to a list of STRINGS for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of FLOATS as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
type(y_axis)
```

```python
# Chain the plot() func to plt and add the axes indide the parenthesis
# Create the plot
plt.plot(x_axis, y_axis)

# recall ,matplotlib.pyplot = plt
```

```python
# want to switch from MATLAB to the object-oriented interface method ,, ie. ude ax.plot instead of plt.plot
```

```python
#To create a figure, use 
fig, ax = plt.subplots()
#To define the axes, use
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.grid()
```

```python
# Also:
# To declare the parameters for line color, width, and marker, we use color=, width=, and marker=, respectively,
# inside the plt.plot() function.
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# ex. 
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# to create a horizontal bar chart use barh()
# # Create the plot
plt.barh(x_axis, y_axis)
```

```python
# to INVERT an axis
# # Create the plot.
plt.barh(x_axis, y_axis, color="magenta")
plt.gca().invert_yaxis()

# gca() means 'Get Current Axis'
```

```python
# VS. invert_xaxis
plt.barh(x_axis, y_axis)
plt.gca().invert_xaxis()
```

```python
# using Object Oriented Approach
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='k', label = 'ny')

# label?
```

```python
# horizontal bar using Obj oriented appr.

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, label = "new york")
```

```python
# If we want the data on opposite axes, we need to switch the arguments in the barh() function
# Create the plot with ax.plt()
fig, ax = plt.subplots()
plt.barh(y_axis, x_axis, label='k')
```
